```python
import numpy as np
import os
import re
import matplotlib.pyplot as plt
```

```python
with open('H2_PES.out') as f:
    s = f.read()
```

```python
e_hf = [float(a) for a in re.findall('energy:\s*([-0-9][0-9.]{2,})', s)]
```

```python
distance = [float(a) for a in re.findall('value:\s*([0-9][0-9.]{2,})', s)]
```

```python
plt.plot(distance[1:20], e_hf[1:20])
plt.xlabel('H-H distance, A')
plt.ylabel('RHF Energy, ha');
```
