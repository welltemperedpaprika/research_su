```python
import numpy as np
import os
import re
import matplotlib.pyplot as plt
```

```python
with open('HF_MP2_aug-cc-pcVDZp_4.out') as f:
    s = f.read()
    
```

```python
e_mp2 = [float(a) for a in re.findall('energy:\s*([-0-9][0-9.]{2,})', s)]
```

```python
distance = [float(a) for a in re.findall('value:\s*([0-9][0-9.]{2,})', s)]
```

```python
plt.plot(distance, e_mp2)
plt.xlabel('H-F distance, A')
plt.ylabel('MP2 Energy, ha');
```
